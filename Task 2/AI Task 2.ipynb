```python
df.fillna(df.select_dtypes(include=['number']).median(), inplace=True)
df.fillna(df.select_dtypes(include=['object']).mode().iloc[0], inplace=True)
```

```python
df = pd.get_dummies(df, columns=['HomePlanet', 'Destination'], drop_first=True)
df['CryoSleep'] = df['CryoSleep'].astype(int)
df['VIP'] = df['VIP'].astype(int)
```

```python
df[['Deck', 'CabinNum', 'Side']] = df['Cabin'].str.split('/', expand=True)
df.drop(columns=['Cabin'], inplace=True)
df['Deck'] = df['Deck'].astype('category').cat.codes
df['Side'] = df['Side'].astype('category').cat.codes
df['CabinNum'] = pd.to_numeric(df['CabinNum'], errors='coerce')
df.fillna(df.median(), inplace=True)  # Fill missing CabinNum values
```
